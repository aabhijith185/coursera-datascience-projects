Question 6
Only looking at the three most populous counties for each state, what are the three most populous states (in order of highest population to lowest population)?
This function should return a list of string values.

```python
def answer_six():
    pop = pf.loc[pf['SUMLEV']==50, ['STNAME', 'CENSUS2010POP']]
    pop = pop.sort_values(['STNAME', 'CENSUS2010POP'],ascending=[True, False]).groupby('STNAME').head(3)
    pop = pop.groupby('STNAME').agg('sum').sort_values('CENSUS2010POP',ascending=False).head(3).reset_index()
    temp = pop.loc[0:2,'STNAME'].tolist()
    return temp
answer_six()             
```

Which county has had the largest change in population within the five year period (hint: population values are stored in columns POPESTIMATE2010 through POPESTIMATE2015, you need to consider all five columns)?
This function should return a single string value.

```python
def answer_seven():
    pop = pf.loc[pf['SUMLEV']==50, ['STNAME', 'CTYNAME', 'NPOPCHG_2011', 'NPOPCHG_2012', 
                                                     'NPOPCHG_2013', 'NPOPCHG_2014', 'NPOPCHG_2015']]
    pop['Change'] = pop['NPOPCHG_2011'] + pop['NPOPCHG_2012'] + pop['NPOPCHG_2013'] + pop['NPOPCHG_2014'] + pop['NPOPCHG_2015']
    return pop.loc[pop['Change'].idxmax(),'CTYNAME']
answer_seven()
```
